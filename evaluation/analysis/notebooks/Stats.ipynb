## Get runtime by average

General idea: Workflows consists of three task levels. For each level, get the max value. The sum of these three values is the runtime.

```python
import pandas as pd

df = pd.read_csv("results-cluster-size-5.csv")

def getDuration(pd):
    start = df['start'].min()
    end = df['end'].max()
    return (end - start)
    
time_extract_data_mean = df['time_extract-data'].mean()
time_core_phrases_mean = df['time_core-phrases'].mean()
time_keywords_mean = df['time_keywords'].mean()
time_stats_mean = df['time_stats'].mean()
time_guess_language_mean = df['time_guess-language'].mean()
time_merger_mean = df['time_merger'].mean()

level2 = [time_core_phrases_mean, time_keywords_mean, time_stats_mean, time_guess_language_mean]

runtime = time_extract_data_mean + max(level2) + time_merger_mean
print(runtime)
df.describe()
```
