# Increasing cluster size at runtime

Starting with a single Pi-Cluster, each 15m another Pi joins the cluster until a cluster of 5 members is reached.
As always, the 60 workflows needs to be processed. Are the joins of a pi visible in the graph? Are the average workflows per Minute reached for each level?

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
colors = ["#2E1510","#5A3343","#5F6483","#399CA5","#64CF94","#E1F36F"]
size=15
params = {'legend.fontsize': 'large',
          'figure.figsize': (10,5),
          'axes.labelsize': size,
          'axes.titlesize': size,
          'xtick.labelsize': size*0.75,
          'ytick.labelsize': size*0.75,
          'axes.titlepad': 25}
plt.rcParams.update(params)

df1 = pd.read_csv("results-cluster-size-1.csv")
df2 = pd.read_csv("results-cluster-size-2.csv")
df3 = pd.read_csv("results-cluster-size-3.csv")
df4 = pd.read_csv("results-cluster-size-4.csv")
df5 = pd.read_csv("results-cluster-size-5.csv")
df_increase = pd.read_csv("results-cluster-size-increasing.csv")

def getWorkflows(pd):
    start = pd['start'].min()
    end = pd['end'].max()
    overallMinutes = (end - start) / 60
    return overallMinutes

durations = [
    getWorkflows(df1),
    getWorkflows(df2),
    getWorkflows(df3),
    getWorkflows(df4),
    getWorkflows(df5),
    getWorkflows(df_increase)
]
print(durations)
categories = ('Cluster 1', 'Cluster 2', 'Cluster 3', 'Cluster 4', 'Cluster 5', "Cluster\nskalierend")
y_pos = np.arange(len(categories))
fig = plt.figure()
plt.barh(y_pos, durations, align='center', color=colors, height=.8)
plt.yticks(y_pos, categories)
plt.xlabel('Laufzeit [m]')
plt.ylabel('Clustergröße')
#plt.title('Laufzeit von 60 Workflows mit unterschiedlichen Clustergrößen')
fig.savefig('cluster_sizes.pdf')

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df1 = pd.read_csv("results-cluster-size-1.csv")
df2 = pd.read_csv("results-cluster-size-2.csv")
df3 = pd.read_csv("results-cluster-size-3.csv")
df4 = pd.read_csv("results-cluster-size-4.csv")
df5 = pd.read_csv("results-cluster-size-5.csv")
df_increase = pd.read_csv("results-cluster-size-increasing.csv")

plt.rcParams['figure.figsize'] = [12, 12]


frames = [df1, df2, df3, df4, df5, df_increase]
interval = np.arange(0, 160, 15)
fig, ax = plt.subplots()

width = 2 # width of bar

plt.axvline(x=15, color="k", linestyle='dashed', linewidth=1) # 2
plt.axvline(x=2*15, color="k", linestyle='dashed', linewidth=1) # 3
plt.axvline(x=3*15, color="k", linestyle='dashed', linewidth=1) # 4
plt.axvline(x=4*15, color="k", linestyle='dashed', linewidth=1) # 5


indexes = np.arange(len(frames))
configs = [
    {
        "color": 'r',
        "label": "df1"
    },
    {
        "color": 'c',
        "label": "df2"
    },
    {
        "color": 'm',
        "label": "df3"
    },
    {
        "color": 'g',
        "label": "df4"
    },
    {
        "color": 'b',
        "label": "df5"
    },
    {
        "color": 'y',
        "label": "Cluster Skalierung"
    }
]

for index, config, df in  zip(indexes, configs, frames):
    #a = df.groupby(pd.cut(df["duration"], interval)).count()['filename']

    min_duration = df["duration"] / 60
    out = pd.cut(min_duration, interval,)
    data = out.value_counts(sort=False)
    dat = np.array(data)
    print(interval)
    #print(dat)
    ax.bar(interval[:-1] + width * index, dat, width, color=config["color"], label=config["label"])
    #ax.set_xticklabels([c[1:-1].replace(","," to") for c in out.cat.categories])

#ax.set_xticklabels(interval)
#ax.autoscale()
plt.xticks(interval)
plt.legend()
plt.show()
```
